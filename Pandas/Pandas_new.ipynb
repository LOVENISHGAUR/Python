<h1>GROUPBY AND AGGREGATE</h1>

Data:
<a href="https://github.com/LOVENISHGAUR/Python/blob/master/Pandas/Data/bikes.csv">
Download File Bike.csv </a>

```python
#Importing Data

bikes = pd.read_csv('bikes.csv', sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
bikes['Berri 1'].plot()
```

<H3>Adding a ‘weekday’ column to our dataframe</h3>

```python
berri_bikes = bikes[['Berri 1']].copy()
berri_bikes[:5]
```

```python
berri_bikes.index
```

```python
berri_bikes.index.day
```

```python
berri_bikes.index.weekday
```

```python
berri_bikes.loc[:,'weekday'] = berri_bikes.index.weekday
berri_bikes[:5]
```

<h3>Adding up the cyclists by weekday</h3>

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind='bar')
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(max)
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts.plot(kind='bar')
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(np.median)
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts.plot(kind='bar')
```
